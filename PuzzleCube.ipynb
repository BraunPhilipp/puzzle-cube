```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from bin import *

# Define Boundaries of Box 4x4
bnd = [4, 4]

# Define XY Rotation Matrix - 4 possible solutions
xy = np.matrix([[0, -1], [1, 0]])

# Define Pieces in x, y Coordinates
a = np.array([[0, 0]])
b = np.array([[1, 0], [1, 1], [2, 1]])
c = np.array([[2, 0], [3, 0], [3, 1], [3, 2], [3, 3]])
d = np.array([[0, 1], [0, 2], [1, 2], [2, 2], [2, 3]])
e = np.array([[0, 3], [1, 3]])

pcs = [a, b, c, d, e]

class Part:
    def __init__(self, n, p):
        self.part_name = n
        self.part_cmbs = []
        self.part_frms = []
        
        self.calc_combinations(p)
        self.part_frms = calc_frame(self.part_cmbs, bnd)
        
    def cmbs(self):
        return self.part_cmbs
        
    def frms(self):
        return self.part_frms
    
    def calc_combinations(self, p):
        prt = []
        # Loop Operation for each piece
        for i in range(4):
            # Rotate each piece 3 times
            p = np.dot(p, xy)
            # Get Rotated Object into inner boundaries
            p = p + np.array(p.min(axis=0) * (-1))

            # Move rotated part
            min_pos = np.squeeze(np.asarray([p.min(axis=0)]))
            max_pos = np.squeeze(np.asarray([p.max(axis=0)]))

            # Index from 0 to 3 (only max - 1)
            for i in range(-min_pos[0], bnd[0]-max_pos[0]):
                for j in range(-min_pos[1], bnd[1]-max_pos[1]):
                    # Create Addition Matrix
                    ad_mat = []
                    for n in range(len(p)):
                        ad_mat = [i, j]
                    # Create new Matrix
                    tmp = np.matrix(p) + np.matrix(ad_mat)
                    # Save temporary computed parts
                    prt.append(np.asarray(tmp)) # np.squeeze(np.asarray(p))

        # Check part for unique positions
        new_prt = [prt[0]]
        for i in range(0,len(prt)):
            for j in range(0, len(new_prt)):
                if np.array_equal(prt[i], new_prt[j]):
                    break;
                if j == ( len(new_prt) - 1 ):
                    new_prt.append(prt[i])
        # Add all unique part combinations
        self.part_cmbs = new_prt

#class Box:
    # array of parts must be saved -> position
    # save all points in box
    
# Create Part Objects     
a = Part("a", pcs[0])
b = Part("b", pcs[1])
c = Part("c", pcs[2])
d = Part("d", pcs[3])
e = Part("e", pcs[4])

# print a.cmbs()[0]
# print a.frms()[0]

# Plot Part
#plt_part = cmb[2][2]

#x_coord = [c[0] for c in plt_part]
#y_coord = [c[1] for c in plt_part]

#plt.plot(x_coord, y_coord, 'rh', linewidth=10)

#plt.ion()
#plt.imshow(cmb[1][0], cmap='gray', interpolation='none') # Correct scaling
```
